```python
import httplib2
import os
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup, SoupStrainer
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import sys
sys.path.append( '../Stocks')
from nse_pickle_fns import *
from dateutil.relativedelta import relativedelta
import pandas as pd
def getSoup(web_url):
    http = httplib2.Http()
    status, response = http.request(web_url)
    return BeautifulSoup(response, "html.parser")
# soup=getSoup(web_url)
# pd.DataFrame(news)
def getHeadlines(url):    
    lis=getSoup(url).find('section',{'id':'pageContent'}).find_all('li')
    return [{'title':l.getText(), 'link':l.find('a').get('href') } for l in lis]
def constructURLs(fromdatefile):
    ordt=datetime.strptime(fromdatefile,'%Y-%m-%d.pkl')

    mhrefs=[]
    while (ordt.toordinal() < datetime.now().toordinal()):
        mhrefs.append(ordt.strftime('/archive/year-%Y,month-%m.cms').replace("-0", "-"))
        ordt=ordt+ relativedelta(months=1)
    return mhrefs
```
